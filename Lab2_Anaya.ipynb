```python
#fn+1 = fn + hG(xn) + O(h2)
#fn+1 = fn + hG(xn +h/2) + O(h3)
    
def euler_and_heun(G, fx0, h, initial, final):
    xs = []
    euler_ys = []
    heun_ys = []

    x = initial
    y_euler = fx0
    y_heun = fx0

    while x <= final:
        xs.append(x)
        euler_ys.append(y_euler)
        heun_ys.append(y_heun)

        # Euler step
        y_euler_next = y_euler + h * G(x)

        # Heun step
        y_heun_next = y_heun + h* G(x + (h/2))

       
        # Update x and y
        x += h
        y_euler = y_euler_next
        y_heun = y_heun_next

    return xs, euler_ys, heun_ys
```
