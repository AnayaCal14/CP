```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#dy/dt = G(t,y)

def RK2(G, y0, h, initial, final):
    ts = []
    ys = []

    t = initial
    y = y0

    while t<=final:
       ts.append(t)
       ys.append(y)
       
       k1 = h * G(t, y)
       k2 = h * G(t +(h/2.0), y + (k1/2.0))
       y_next = y + k2
#update 
       t += h
       y = y_next

    return ts, ys

def RK4(G, y0, h, initial, final):
    xs = []
    fx_s = []
    x = initial 
    fx = y0

    while x<=final:
       xs.append(x)
       fx_s.append(fx)
       
       k1 = h * G(x, fx)
       k2 = h * G(x +(h/2.0), fx + (k1/2.0))
       k3 = h * G(x +(h/2.0), fx + (k2/2.0))
       k4 = h * G(x + h, fx + k3)
       fx_next = fx + ((1/3)*((k1/2.0) + k2 + k3 +(k4/2.0)))
    #update 
       x += h
       fx = fx_next

    return xs, fx_s
 
```

```python
#1 
def G(t, y):
    fn = y - (t**2) + 1
    return fn
    
T, Y = RK2(G, 0.5, 0.1, 0, 2)
T = np.array(T)

plt.plot(T, Y, label='RK2', linestyle='--', color = 'red')
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Runge-Kutta method")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# analytic 
for t in range (0, 2):

 
 Fx = (T**2) + (2*T) + 1 - 0.5*(np.exp(T))

plt.plot(T, Y, label='RK2', linestyle='--')
plt.plot(T, Fx, label='Analytic', linestyle='-')
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Runge-Kutta method")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ERROR FUNCTION 
plt.plot(T, np.abs(Fx - Y), label='error')
plt.xlabel("t")
plt.ylabel("error")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2 

def G(x, y):
    return ((x+y)*np.sin(x))

X, Y = RK2(G, 5, 0.1, 0, 6)
X = np.array(X)

X_prime, Fx = RK4(G, 5, 0.1, 0, 6)
X_prime = np.array(X_prime)

plt.plot(X, Y, label='RK2', linestyle='--')
plt.plot(X_prime, Fx, label='RK4', linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Runge-Kutta method")
plt.legend()
plt.grid(True)
plt.show()
```
