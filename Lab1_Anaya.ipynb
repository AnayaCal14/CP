```python
#Q1
def quad_root(a, b, c):
    D = b**2 - (4*a*c)

    if D > 0:
        root1 = (-b + np.sqrt(D)) / (2*a)
        root2 = (-b - np.sqrt(D)) / (2*a)
        print("D > 0: Two real and distinct roots")
    elif D == 0:
        root1 = root2 = -b / (2*a)
        print("D = 0: One real root (repeated)")
    else:
        real_part = -b / (2*a)
        imag_part = np.sqrt(-D) / (2*a)
        root1 = complex(real_part, imag_part)
        root2 = complex(real_part, -imag_part)
    return root1, root2
```

```python
print (quad_root(1, -2, 1))
```

```python
#Q2
def inverse_matrix(a, b, c, d):
    det = (a*d) -(b*c)
    if det == 0:
        print("singular matrix and non-invertible")
    else:
        inverse = np.array([[d, -b], [-c, a]])/ det
        
    return inverse 
```

```python
print("the inverse of the matrix [[1, 2],[3, 4]] is", inverse_matrix(1, 2, 3, 4))
```

```python
#Q3
# ax + by = c and px + qy = r 
def two_eqn_sol(a, b, c, p, q, r):
    x = ((c*q) - (b*r))/((a*q) - (b*p))
    y = ((c*p) - (a*r))/((b*p) - (a*q))

    return x,y 
```

```python
print("Solutions of the equations are (x,y) =", two_eqn_sol(1, 2, 3, 4, 5, 6))
```

```python
#Q4
def sin_taylor(x, terms=10):
    result = 0
    for n in range(terms):
        term = ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
        result += term
    return result


sin_taylor_vec = np.vectorize(sin_taylor)


h1 = np.pi / 20
h2 = np.pi / 3


x1 = np.arange(0, 2*np.pi, h1)
x2 = np.arange(0, 2*np.pi, h2)

# Calculate values
y1 = sin_taylor_vec(x1, terms=10)
y2 = sin_taylor_vec(x2, terms=10)

# True sin(x) for reference
x_fine = np.linspace(0, 2*np.pi, 500)
y_true = np.sin(x_fine)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_fine, y_true, label='sin(x)', color='blue')
plt.plot(x1, y1, 'ro--', label='Taylor Approx. (h = /20)')
plt.title('Taylor Series Approximation of sin(x) with Different Step Sizes')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig
plt.plot(x_fine, y_true, label='sin(x)', color='blue')
plt.plot(x2, y2, 'gs--', label='Taylor Approx. (h = /3)')
plt.title('Taylor Series Approximation of sin(x) with Different Step Sizes')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
plt.savefig
```

```python
#Q5
numbers = [124,45,89,67,30,200,145,765,18,0,11]
#Ascending = sorted(numbers)
#Descending = sorted(numbers, reverse=True)
def Ascending(arr):
    n=len(arr)
    for i in range (n-1):
        swapped = False  # Optimization: flag to check if any swap occurred in a pass
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no two elements were swapped by inner loop, then the array is sorted
        if not swapped:
            break
    return arr

def Descending(arr):
    n=len(arr)
    for i in range (n-1):
        swapped = False  # Optimization: flag to check if any swap occurred in a pass
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no two elements were swapped by inner loop, then the array is sorted
        if not swapped:
            break
    return arr
    
print (Ascending(numbers))
print(Descending(numbers))
```

```python
#Q6
def Factorial(N):
    if N < 0:
        return print("factorial not defined")
    elif N == 0:
        return 1
    else:
        return math.gamma(N+1)
```
